```python
import matplotlib.pyplot as plt
import numpy as np


# Define the function H(X)
def H(p):
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)


# Create an array of p values from 0 to 1
p = np.linspace(0, 1, 500)

# Compute H(p) and handle the cases where p is 0 or 1
H_values = np.where((p == 0) | (p == 1), 0, H(p))

# Create the plot
plt.figure(figsize=(6, 6))
plt.plot(p, H_values, 'b-', linewidth=2)
plt.xlabel('Pr(X = 1)')
plt.ylabel('H(X)')
plt.grid(True)
plt.xlim(0, 1.008)
plt.ylim(0, 1.008)

plt.title('Entropy function', fontsize=14, pad=10)

plt.show()
```
